```python
interessanteCols = ["preis","anzahl_raeume","wohnflaeche","transaktionsArt","url","datumDownload","plz"]
```

```python
def make_googlemap_link(df,startzip="22846"):
    googlempadir_root = 'https://www.google.de/maps/dir/'
    root_and_start = googlempadir_root + startzip+'/'
    googlemap_url = df.apply(
        lambda row: root_and_start + row['plz']
        ,axis=1)
    loc_url = df.columns.get_loc("url")
    test = df.copy()
    test.insert(loc_url+1,"url_maps",googlemap_url)
    return test
make_googlemap_link(df)
                  
```
